Importiamo tutti i moduli necessari

```python
import statistics as st
import modin.pandas as pd
import numpy as np
from sklearn.preprocessing import scale
```

Una volta aver importato i moduli, si passa al calcolo del **minimo**,**massimo**,**valore medio**,**deviazione standard**,**varianza** e **valore totale** delle feature numeriche contenute nel data set. Questo Ã¨ possibile grazie alle funzioni contenute nel modulo **statistics**.

```python
stDeviations = []
min_ = []
max_ = []
variance = []
mean = []
sums = []
#author_ranking = calculateAuthorRanking()
table = pd.read_excel('..\\data\\TabellaCompleta.xlsx') 
#df2 = pd.read_excel('C:\\Users\\utente\\Desktop\\PanelJam\\PanelJamDataExtractor\\TabellaProgettiPanelJam.xlsx')

#Calcolo deviazioni standard features numeriche
#stars
stars_attuale = table['stars_attuale'].tolist()
stDeviations.append(st.stdev(stars_attuale))


#followers
followers_attuale = table['followers_attuale'].tolist()
stDeviations.append(st.stdev(followers_attuale))


#totLoves
tot_loves_attuale = table['tot_loves_attuale'].tolist()
stDeviations.append(st.stdev(tot_loves_attuale))

#Tot views
tot_views_attuale = table['tot_views_attuale'].tolist()
stDeviations.append(st.stdev(tot_views_attuale))


#authors ranking
ranking_attuale = table['ranking_attuale'].tolist()
stDeviations.append(st.stdev(ranking_attuale))

#likes
likes = table['Likes'].tolist()
stDeviations.append(st.stdev(likes))

#views
views = table['Views'].tolist()
stDeviations.append(st.stdev(views))

#comments
comments = table['Comments'].tolist()
stDeviations.append(st.stdev(comments))

#Project depth
project_depth = table['project_depth'].tolist()
stDeviations.append(st.stdev(project_depth))

#time
time = table['time'].tolist()
numTime = list(map(lambda sub:int(''.join([ele for ele in sub if ele.isnumeric()])), time))
stDeviations.append(st.stdev(numTime))


#Calcolo valori minimi
min_.append(min(stars_attuale))
min_.append(min(followers_attuale))
min_.append(min(tot_loves_attuale))
min_.append(min(tot_views_attuale))
min_.append(min(ranking_attuale))
min_.append(min(likes))
min_.append(min(views))
min_.append(min(comments))
min_.append(min(project_depth))
min_.append(min(numTime))

#Calcolo valori massimi
max_.append(max(stars_attuale))
max_.append(max(followers_attuale))
max_.append(max(tot_loves_attuale))
max_.append(max(tot_views_attuale))
max_.append(max(ranking_attuale))
max_.append(max(likes))
max_.append(max(views))
max_.append(max(comments))
max_.append(max(project_depth))
max_.append(max(numTime))

#Calcolo valori medi
mean.append(st.mean(stars_attuale))
mean.append(st.mean(followers_attuale))
mean.append(st.mean(tot_loves_attuale))
mean.append(st.mean(tot_views_attuale))
mean.append(st.mean(ranking_attuale))
mean.append(st.mean(likes))
mean.append(st.mean(views))
mean.append(st.mean(comments))
mean.append(st.mean(project_depth))
mean.append(st.mean(numTime))

#Calcolo varianze

variance.append(st.variance(stars_attuale))
variance.append(st.variance(followers_attuale))
variance.append(st.variance(tot_loves_attuale))
variance.append(st.variance(tot_views_attuale))
variance.append(st.variance(ranking_attuale))
variance.append(st.variance(likes))
variance.append(st.variance(views))
variance.append(st.variance(comments))
variance.append(st.variance(project_depth))
variance.append(st.variance(numTime))

#Calcolo somme

sums.append(sum(stars_attuale))
sums.append(sum(followers_attuale))
sums.append(sum(tot_loves_attuale))
sums.append(sum(tot_views_attuale))
sums.append(sum(ranking_attuale))
sums.append(sum(likes))
sums.append(sum(views))
sums.append(sum(comments))
sums.append(sum(project_depth))
sums.append(sum(numTime))
```

Una volta calcolati tutti i valori necessari si procede con la costruzione della tabella(mostrata di seguito) utilizzando il modulo **modin**.

```python
data = {'Feature': ['stars_attuale','followers_attuale','tot_loves_attuale','tot_views_attuale','ranking_attuale','likes','views','comments','project_depth','time'],
            'Total': sums,
            'Min':  min_,
            'Max': max_,
            'Mean': mean,
            'St. Dev.': stDeviations,
            'Variance': variance
            }

df = pd.DataFrame(data, columns = ['Feature','Total','Min','Max','Mean','St. Dev.','Variance'])
#df.to_excel('numericalStats.xlsx', index=False)
df
```
